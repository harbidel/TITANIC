```python
train = train[['PassengerId','Pclass', 'Sex', 'Age','SibSp','Fare','Parch','Embarked','Survived']]
train.head()
```

```python
test = test[['PassengerId','Pclass', 'Sex', 'Age','SibSp','Fare','Parch','Embarked']]
test.head()
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
train["Sex"] = lb_make.fit_transform(train["Sex"])
train[["Sex"]].head()
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
train["Embarked"] = lb_make.fit_transform(train["Embarked"])
train[["Sex","Embarked"]].head()
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
test["Sex"] = lb_make.fit_transform(test["Sex"])
test[["Sex"]].head()
```

```python
from sklearn.preprocessing import LabelEncoder

lb_make = LabelEncoder()
test["Embarked"] = lb_make.fit_transform(test["Embarked"])
test[["Sex","Embarked"]].head()
```

```python
train = train.fillna(train.mean())
train.head()
```

```python
train.info()
```

```python
test.head()
```

```python
test = test.fillna(test.mean())
test.head()
```

```python
test.info()
```

```python
X_train = ['Pclass', 'Sex', 'Age','SibSp','Parch','Fare','Embarked']
y_train = 'Survived'
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
classifier.fit(train[X_train], train[y_train])
```

```python
X_test = ['Pclass', 'Sex', 'Age','SibSp','Parch','Fare','Embarked']
X_test
```

```python
y_pred = classifier.predict(test[X_test])
y_pred
```

```python
#Create a  DataFrame with the passengers ids and our prediction regarding whether they survived or not
submission = pd.DataFrame({'PassengerId':test['PassengerId'],'Survived':y_pred})
submission.head()
```

```python
filename = 'My Titanic Predictions 7.csv'

submission.to_csv(filename,index=False)

print('Saved file: ' + filename)
```

```python
from sklearn.tree import DecisionTreeClassifier

#Create classifier object with default hyperparameters
clf = DecisionTreeClassifier()  

#Fit our classifier using the training features and the training target values
clf.fit(train[X_train],train[y_train]) 
```

```python
from sklearn.tree import export_graphviz
export_graphviz(clf,out_file='titanic_tree.dot',feature_names=X_train,rounded=True,filled=True,class_names=['Survived','Did not Survive'])
```

```python
predictions = clf.predict(test[X_test])

#Display our predictions - they are either 0 or 1 for each training instance 
#depending on whether our algorithm believes the person survived or not.
predictions
```

```python
submission2 = pd.DataFrame({'PassengerId':test['PassengerId'],'Survived':predictions})

#Visualize the first 5 rows
submission2.head()
```
